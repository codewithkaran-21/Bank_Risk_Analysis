### Risk Analysis - Overview
**Risk Weighting** is our target variable. It represents the level of financial risk associated with each client.
# - 0 = Low Risk
# - 1 = Medium Risk  
# - 2 = Medium-High Risk
# - 3 = High Risk
# - 4+ = Very High Risk

```python
# Check Risk Distribution
print("Risk Weighting Distribution:")
print(df['Risk Weighting'].value_counts().sort_index())
print("\n")
print("Risk Distribution (%):")
print(df['Risk Weighting'].value_counts(normalize=True).sort_index()*100)
```

```python
# Visualize Risk Distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Count plot
df['Risk Weighting'].value_counts().sort_index().plot(kind='bar', ax=axes[0], color='steelblue')
axes[0].set_title('Risk Weighting Distribution (Count)', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Risk Level')
axes[0].set_ylabel('Number of Clients')

# Pie chart
risk_counts = df['Risk Weighting'].value_counts().sort_index()
axes[1].pie(risk_counts, labels=risk_counts.index, autopct='%1.1f%%', startangle=90)
axes[1].set_title('Risk Weighting Distribution (%)', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Numerical Risk Analysis
# Analyze how numerical features relate to Risk
numerical_cols = ['Age', 'Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 
                  'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts']

print("Mean values by Risk Weighting:\n")
risk_summary = df.groupby('Risk Weighting')[numerical_cols].mean()
print(risk_summary)
```

```python
# Box plots for continuous variables vs Risk
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.ravel()

features = ['Age', 'Estimated Income', 'Bank Loans', 'Bank Deposits']

for idx, feature in enumerate(features):
    sns.boxplot(data=df, x='Risk Weighting', y=feature, ax=axes[idx], palette='Set2',legend=False)
    axes[idx].set_title(f'{feature} vs Risk Weighting', fontweight='bold')
    axes[idx].set_xlabel('Risk Level')

plt.tight_layout()
plt.show()
```

```python
# Categorical Risk Analysis - Gender
print("Risk Distribution by Gender:")
risk_by_gender = pd.crosstab(df['GenderId'], df['Risk Weighting'], margins=True)
print(risk_by_gender)
print("\n")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Count plot
sns.countplot(data=df, x='Risk Weighting', hue='GenderId', ax=axes[0], palette='Set2',legend=False)
axes[0].set_title('Risk Distribution by Gender', fontweight='bold')
axes[0].set_xlabel('Risk Level')
axes[0].set_ylabel('Count')

# Percentage distribution
risk_pct = pd.crosstab(df['GenderId'], df['Risk Weighting'], normalize='index') * 100
risk_pct.T.plot(kind='bar', ax=axes[1], color=['skyblue', 'lightcoral'])
axes[1].set_title('Risk Distribution by Gender (%)', fontweight='bold')
axes[1].set_xlabel('Risk Level')
axes[1].set_ylabel('Percentage')
axes[1].legend(title='Gender', labels=['Gender 1', 'Gender 2', 'Gender 3'])

plt.tight_layout()
plt.show()
```

```python
# Risk Analysis by Fee Structure
print("Risk Distribution by Fee Structure:")
risk_by_fee = pd.crosstab(df['Fee Structure'], df['Risk Weighting'], margins=True)
print(risk_by_fee)
print("\n")

fig, ax = plt.subplots(figsize=(12, 5))
sns.countplot(data=df, x='Fee Structure', hue='Risk Weighting', ax=ax, palette='husl')
ax.set_title('Risk Weighting by Fee Structure', fontweight='bold', fontsize=12)
ax.set_xlabel('Fee Structure')
ax.set_ylabel('Number of Clients')
plt.tight_layout()
plt.show()
```

```python
# Correlation Analysis - Numerical Features vs Risk
fig, ax = plt.subplots(figsize=(12, 8))

numerical_features = ['Age', 'Estimated Income', 'Superannuation Savings', 'Amount of Credit Cards',
                     'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 
                     'Saving Accounts', 'Foreign Currency Account', 'Business Lending', 'Properties Owned']

correlation_with_risk = df[numerical_features + ['Risk Weighting']].corr()['Risk Weighting'].sort_values(ascending=False)
print("Correlation with Risk Weighting:")
print(correlation_with_risk)
print("\n")

correlation_with_risk[:-1].plot(kind='barh', ax=ax, color='steelblue')
ax.set_title('Feature Correlation with Risk Weighting', fontweight='bold', fontsize=12)
ax.set_xlabel('Correlation Coefficient')
plt.tight_layout()
plt.show()
```

```python
# Heatmap of Correlations
fig, ax = plt.subplots(figsize=(14, 10))
correlation_matrix = df[numerical_features + ['Risk Weighting']].corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0, ax=ax, cbar_kws={'label': 'Correlation'})
ax.set_title('Correlation Matrix - All Numerical Features', fontweight='bold', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Risk Segmentation Analysis - Income Band vs Risk
print("Risk Distribution by Income Band:")
risk_by_income = pd.crosstab(df['Income Band'], df['Risk Weighting'], margins=True)
print(risk_by_income)
print("\n")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Count plot
sns.countplot(data=df, x='Income Band', hue='Risk Weighting', ax=axes[0], palette='viridis')
axes[0].set_title('Risk Distribution by Income Band', fontweight='bold')
axes[0].set_xlabel('Income Band')
axes[0].set_ylabel('Count')

# Stacked bar chart
risk_income_pct = pd.crosstab(df['Income Band'], df['Risk Weighting'], normalize='index') * 100
risk_income_pct.plot(kind='bar', stacked=False, ax=axes[1], colormap='viridis')
axes[1].set_title('Risk Distribution by Income Band (%)', fontweight='bold')
axes[1].set_xlabel('Income Band')
axes[1].set_ylabel('Percentage')
axes[1].legend(title='Risk Level', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# Age Group Analysis
df['Age Group'] = pd.cut(df['Age'], bins=[0, 25, 35, 50, 65, 100], 
                         labels=['18-25', '26-35', '36-50', '51-65', '65+'])

print("Risk Distribution by Age Group:")
risk_by_age = pd.crosstab(df['Age Group'], df['Risk Weighting'], margins=True)
print(risk_by_age)
print("\n")

fig, ax = plt.subplots(figsize=(12, 5))
sns.countplot(data=df, x='Age Group', hue='Risk Weighting', ax=ax, palette='RdYlGn_r')
ax.set_title('Risk Weighting by Age Group', fontweight='bold', fontsize=12)
ax.set_xlabel('Age Group')
ax.set_ylabel('Number of Clients')
plt.tight_layout()
plt.show()
```

```python
# Credit Card Analysis - Potential Risk Indicator
print("Risk Distribution by Number of Credit Cards:")
print(pd.crosstab(df['Amount of Credit Cards'], df['Risk Weighting'], margins=True))
print("\n")

# Average Credit Card Balance by Risk
print("Average Credit Card Balance by Risk Level:")
print(df.groupby('Risk Weighting')['Credit Card Balance'].describe())
print("\n")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Credit cards count
sns.countplot(data=df, x='Amount of Credit Cards', hue='Risk Weighting', ax=axes[0], palette='cool')
axes[0].set_title('Risk by Number of Credit Cards', fontweight='bold')
axes[0].set_xlabel('Amount of Credit Cards')
axes[0].set_ylabel('Count')

# Credit card balance distribution
sns.violinplot(data=df, x='Risk Weighting', y='Credit Card Balance', ax=axes[1], palette='Set2')
axes[1].set_title('Credit Card Balance Distribution by Risk', fontweight='bold')
axes[1].set_xlabel('Risk Level')
axes[1].set_ylabel('Credit Card Balance')

plt.tight_layout()
plt.show()
```

```python
# Loan to Deposit Ratio Analysis - Key Risk Metric
df['Loan_to_Deposit_Ratio'] = df['Bank Loans'] / (df['Bank Deposits'] + 1)  # +1 to avoid division by zero

print("Loan to Deposit Ratio by Risk Level:")
print(df.groupby('Risk Weighting')['Loan_to_Deposit_Ratio'].describe())
print("\n")

fig, ax = plt.subplots(figsize=(12, 5))
sns.boxplot(data=df, x='Risk Weighting', y='Loan_to_Deposit_Ratio', ax=ax, palette='coolwarm')
ax.set_title('Loan to Deposit Ratio by Risk Weighting', fontweight='bold', fontsize=12)
ax.set_xlabel('Risk Level')
ax.set_ylabel('Loan to Deposit Ratio')
plt.tight_layout()
plt.show()
```

```python
# Customer Tenure Analysis
df['Tenure_Years'] = (pd.to_datetime('2024-02-03') - pd.to_datetime(df['Joined Bank'], format='%d-%m-%Y')).dt.days / 365.25

print("Tenure (Years) by Risk Level:")
print(df.groupby('Risk Weighting')['Tenure_Years'].describe())
print("\n")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Box plot
sns.boxplot(data=df, x='Risk Weighting', y='Tenure_Years', ax=axes[0], palette='Set1')
axes[0].set_title('Customer Tenure by Risk Weighting', fontweight='bold')
axes[0].set_xlabel('Risk Level')
axes[0].set_ylabel('Tenure (Years)')

# Violin plot
sns.violinplot(data=df, x='Risk Weighting', y='Tenure_Years', ax=axes[1], palette='muted')
axes[1].set_title('Tenure Distribution by Risk', fontweight='bold')
axes[1].set_xlabel('Risk Level')
axes[1].set_ylabel('Tenure (Years)')

plt.tight_layout()
plt.show()
```

```python
# Nationality Risk Profile
print("Risk Distribution by Nationality:")
risk_by_nationality = pd.crosstab(df['Nationality'], df['Risk Weighting'], margins=True)
print(risk_by_nationality)
print("\n")

# Top nationalities
top_nationalities = df['Nationality'].value_counts().head(5).index
df_top_nat = df[df['Nationality'].isin(top_nationalities)]

fig, ax = plt.subplots(figsize=(12, 5))
sns.countplot(data=df_top_nat, x='Nationality', hue='Risk Weighting', ax=ax, palette='Set3')
ax.set_title('Risk Weighting by Top 5 Nationalities', fontweight='bold', fontsize=12)
ax.set_xlabel('Nationality')
ax.set_ylabel('Number of Clients')
plt.tight_layout()
plt.show()
```

```python
# Risk Score Calculation - Composite Risk Index
# Creating a risk score based on multiple factors
df['Risk_Score'] = 0

# Age factor: Younger and older people are higher risk
df.loc[df['Age'] < 30, 'Risk_Score'] += 1
df.loc[df['Age'] > 70, 'Risk_Score'] += 1

# High credit card balance
df.loc[df['Credit Card Balance'] > df['Credit Card Balance'].quantile(0.75), 'Risk_Score'] += 1

# High loan to deposit ratio
df.loc[df['Loan_to_Deposit_Ratio'] > df['Loan_to_Deposit_Ratio'].quantile(0.75), 'Risk_Score'] += 2

# Low income
df.loc[df['Estimated Income'] < df['Estimated Income'].quantile(0.25), 'Risk_Score'] += 1

# Multiple credit cards
df.loc[df['Amount of Credit Cards'] > 2, 'Risk_Score'] += 1

# Low tenure
df.loc[df['Tenure_Years'] < 3, 'Risk_Score'] += 1

# High business lending
df.loc[df['Business Lending'] > df['Business Lending'].quantile(0.75), 'Risk_Score'] += 1

print("Calculated Risk Score Distribution:")
print(df['Risk_Score'].value_counts().sort_index())
print("\n")

# Compare with actual risk weighting
print("Correlation between Calculated Risk Score and Actual Risk Weighting:")
print(f"Correlation: {df['Risk_Score'].corr(df['Risk Weighting']):.4f}")
print("\n")

fig, ax = plt.subplots(figsize=(12, 5))
sns.scatterplot(data=df, x='Risk_Score', y='Risk Weighting', ax=ax, alpha=0.6, s=50)
ax.set_title('Calculated Risk Score vs Actual Risk Weighting', fontweight='bold', fontsize=12)
ax.set_xlabel('Calculated Risk Score')
ax.set_ylabel('Actual Risk Weighting')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Client Segmentation by Risk Profile
risk_profiles = pd.DataFrame({
    'Risk Level': df['Risk Weighting'].unique(),
    'Client Count': [len(df[df['Risk Weighting'] == x]) for x in sorted(df['Risk Weighting'].unique())],
})
risk_profiles['Avg Income'] = [df[df['Risk Weighting'] == x]['Estimated Income'].mean() 
                               for x in sorted(df['Risk Weighting'].unique())]
risk_profiles['Avg Age'] = [df[df['Risk Weighting'] == x]['Age'].mean() 
                           for x in sorted(df['Risk Weighting'].unique())]
risk_profiles['Avg Tenure'] = [df[df['Risk Weighting'] == x]['Tenure_Years'].mean() 
                              for x in sorted(df['Risk Weighting'].unique())]

print("Risk Profile Summary:")
print(risk_profiles)
print("\n")
```

```python
# Create a comprehensive Risk Dashboard
fig = plt.figure(figsize=(16, 10))
gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)

# 1. Risk distribution
ax1 = fig.add_subplot(gs[0, 0])
df['Risk Weighting'].value_counts().sort_index().plot(kind='bar', ax=ax1, color='steelblue')
ax1.set_title('Risk Distribution', fontweight='bold')
ax1.set_xlabel('Risk Level')
ax1.set_ylabel('Count')

# 2. Age vs Risk
ax2 = fig.add_subplot(gs[0, 1])
df.boxplot(column='Age', by='Risk Weighting', ax=ax2)
ax2.set_title('Age by Risk Level', fontweight='bold')
ax2.get_figure().suptitle('')

# 3. Income vs Risk
ax3 = fig.add_subplot(gs[0, 2])
df.boxplot(column='Estimated Income', by='Risk Weighting', ax=ax3)
ax3.set_title('Income by Risk Level', fontweight='bold')
ax3.get_figure().suptitle('')

# 4. Loyalty vs Risk
ax4 = fig.add_subplot(gs[1, 0])
sns.countplot(data=df, x='Loyalty Classification', hue='Risk Weighting', ax=ax4, palette='Set2')
ax4.set_title('Loyalty vs Risk', fontweight='bold')
ax4.set_xlabel('Loyalty')
plt.setp(ax4.xaxis.get_majorticklabels(), rotation=45)

# 5. Fee Structure vs Risk
ax5 = fig.add_subplot(gs[1, 1])
sns.countplot(data=df, x='Fee Structure', hue='Risk Weighting', ax=ax5, palette='Set1')
ax5.set_title('Fee Structure vs Risk', fontweight='bold')

# 6. Loan to Deposit Ratio
ax6 = fig.add_subplot(gs[1, 2])
df.boxplot(column='Loan_to_Deposit_Ratio', by='Risk Weighting', ax=ax6)
ax6.set_title('Loan/Deposit Ratio vs Risk', fontweight='bold')
ax6.get_figure().suptitle('')

# 7. Credit Cards
ax7 = fig.add_subplot(gs[2, 0])
sns.countplot(data=df, x='Amount of Credit Cards', hue='Risk Weighting', ax=ax7, palette='husl')
ax7.set_title('Credit Cards vs Risk', fontweight='bold')

# 8. Properties Owned
ax8 = fig.add_subplot(gs[2, 1])
sns.countplot(data=df, x='Properties Owned', hue='Risk Weighting', ax=ax8, palette='coolwarm')
ax8.set_title('Properties vs Risk', fontweight='bold')

# 9. Risk Score
ax9 = fig.add_subplot(gs[2, 2])
df.boxplot(column='Risk_Score', by='Risk Weighting', ax=ax9)
ax9.set_title('Calculated Risk Score vs Actual', fontweight='bold')
ax9.get_figure().suptitle('')

fig.suptitle('Bank Risk Analysis Dashboard', fontsize=16, fontweight='bold', y=0.995)
plt.show()
```
